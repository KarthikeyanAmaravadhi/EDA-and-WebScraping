```python
productname = []
rating = []
price = []
features = []
pagenum = []
discount = []


for i in range(1,21):
    start_time = time.time()
    URL = 'https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}'.format(i)
    page=requests.get(URL)
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)
    
    for x in soup.find_all('div',attrs = {'class':'_3pLy-c row'}):
        pname = x.find('div',attrs = {'class':'_4rR01T'})
        cost = x.find('div',attrs ={'class':'_1_WHN1'})
        rat = x.find('div',attrs = {'class':'_3LWZlK'})
        specs = x.find('ul',attrs = {'class':'_1xgFaf'})
        dis = x.find('div',attrs = {'class':'_3Ay6Sb'})
        
        
        if pname is None:
            productname.append(np.NaN)
        else:
            productname.append(pname.text)
        
        if cost is None:
            price.append(np.NaN)
        else:
            price.append(cost.text)
            
        if rat is None:
            rating.append(np.NaN)
        else:
            rating.append(rat.text)
        
        if specs is None:
            features.append(np.NaN)
        else:
            features.append(specs.text)
        if dis is None:
            discount.append(np.NaN)
        else:
            discount.append(dis.text)
        
        pagenum.append(i)
            
    print('Page {} completed in {} seconds'.format(i,time.time()-start_time))
```

## Converting into DataFrame

```python
flipkart_df = pd.DataFrame({'Product':productname,'Price':price,'Features':features,'Rating':rating,'PageNo':pagenum,'Discount':discount})
```

## Data Extraction

```python
flipkart_df.head()
```

```python
regex = r'[\d.\d]+\sinch'
flipkart_df['Display_details'] = flipkart_df['Features'].apply(lambda x: re.compile(regex).findall(x))
```

```python
flipkart_df.head()
```

```python
regex = r'[Intel|AMD][\sA-Za-z0-9]+Processor'
```

```python
flipkart_df['Processor'] = flipkart_df['Features'].apply(lambda x: re.compile(regex).findall(x))
```

```python
flipkart_df.head()
```

```python
regex = r"[\d\]\s[GB]+\s[A-Z0-9]+\sRAM"
flipkart_df['Memory'] = flipkart_df['Features'].apply(lambda x: re.compile(regex).findall(x))
```

```python
flipkart_df.head()
```

```python
regex = r'[0-9\s]+(?:GB|TB)\s(?:HDD|SSD)'
flipkart_df ['Storage']= flipkart_df['Features'].apply(lambda x: re.compile(regex).findall(x))
```

```python
flipkart_df.head()
```

```python
regex = r'^\w+'
flipkart_df['Brand'] = flipkart_df['Product'].apply(lambda x : re.compile(regex).findall(x))
```

```python
flipkart_df.head()
```

```python
flipkart_df.info()
```

```python
flipkart_df.head(10)
```

```python
regex = r'[\d+\s]+bit[\s\w]+Operating System'
flipkart_df['Bit&OS'] = flipkart_df['Features'].apply(lambda x :re.compile(regex).findall(x))
```

```python
flipkart_df.head()
```

```python
flipkart_df['Processor'] = flipkart_df['Processor'].apply(lambda x :''.join(x))
flipkart_df['Memory'] = flipkart_df['Memory'].apply(lambda x :''.join(x))
flipkart_df['Storage'] = flipkart_df['Storage'].apply(lambda x :''.join(x))
flipkart_df['Brand'] = flipkart_df['Brand'].apply(lambda x :''.join(x))
flipkart_df['Display_details'] = flipkart_df['Display_details'].apply(lambda x :''.join(x))
flipkart_df['Bit&OS'] = flipkart_df['Bit&OS'].apply(lambda x : ''.join(x))
```

```python
flipkart_df['Price'] = flipkart_df['Price'].apply(lambda x :x.replace('₹','').replace(',','')).astype(float)
```

```python
flipkart_df.head()
```

```python
flipkart_df['Discount'] = flipkart_df['Discount'].str.replace('\%','')
```
